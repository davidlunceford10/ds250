In VS Code, the keyboard shortcuts for adding new cells in Jupyter notebooks are:

- **`B`** - Add new cell below the current cell
- **`A`** - Add new cell above the current cell
- **`D`** - Click twice to delete current cell

These work when you're in **command mode** (press `Esc` to enter command mode if you're in edit mode).

If you prefer using traditional keyboard shortcuts:
- **`Ctrl+Shift+Alt+N`** (Windows/Linux) or **`Cmd+Shift+Option+N`** (Mac) - Insert cell below

You can also customize these shortcuts in VS Code by going to `File > Preferences > Keyboard Shortcuts` and searching for "notebook.cell.insertCodeCellBelow" or "notebook.cell.insertCodeCellAbove".
